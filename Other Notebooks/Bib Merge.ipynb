```python
def read_bib_file(file_path):
    with open(file_path) as bibtex_file:
        bib_database = bibtexparser.load(bibtex_file)
    return bib_database

def get_doi_set(bib_database):
    return {entry.get('doi', '').lower() for entry in bib_database.entries}

def filter_bib_database(source_db, filter_dois):
    return [entry for entry in source_db.entries if entry.get('doi', '').lower() in filter_dois]

def write_bib_file(file_path, filtered_entries):
    with open(file_path, 'w') as bibtex_file:
        new_db = bibtexparser.bparser.BibDatabase()
        new_db.entries = filtered_entries
        bibtexparser.dump(new_db, bibtex_file)

def count_entries_in_file(file_path):
    bib_database = read_bib_file(file_path)
    return len(bib_database.entries)
```

```python
# Download savedrecs.bib file
!wget https://raw.githubusercontent.com/5H5KN5/SIT723/main/Impact/Bib/Merge/savedrecs_merge.bib

# Download screened.bib file
!wget https://raw.githubusercontent.com/5H5KN5/SIT723/main/Impact/Bib/Screened/screened.bib
```

```python
# File paths
merged_bib_path = 'savedrecs_merge.bib'
screened_bib_path = 'screened.bib'
output_bib_path = 'screened_recs.bib'

# Read the .bib files
merged_db = read_bib_file(merged_bib_path)
screened_db = read_bib_file(screened_bib_path)

# Count the number of entries in the merged and screened files
merged_count = count_entries_in_file(merged_bib_path)
screened_count = count_entries_in_file(screened_bib_path)
print(f"Number of entries in merged.bib: {merged_count}")
print(f"Number of entries in screened.bib: {screened_count}")

# Extract DOIs
screened_dois = get_doi_set(screened_db)

# Filter merged database
filtered_entries = filter_bib_database(merged_db, screened_dois)

# Write the filtered database to a new .bib file
write_bib_file(output_bib_path, filtered_entries)

# Count the number of entries in the output file
output_count = count_entries_in_file(output_bib_path)
print(f"Number of entries in Screened_Merged.bib: {output_count}")
```
